## Interactive Visualizations

## Histogram

```python
data['Shipping Cost'].iplot(kind='hist')
```

```python
data['Profit'].iplot(kind='hist')
```

## Scatter plots

```python
data.iplot(kind='scatter', x='Discount', y='Profit', mode='markers', size=7)
```

```python
data.iplot(kind='scatter', x='Shipping Cost', y='Profit', mode='markers', size=5,
             xTitle='Shipping Cost', yTitle='Profit', 
              title='Shipping Cost v/s Profit')
```

```python
# adding a third category
fig = px.scatter(data,x='Discount',y='Profit',color='Category',
                title='Discount v/s Profit|Category')
fig.show()
```

```python
fig = px.scatter(data,x='Shipping Cost',y='Profit',color='Category',
                title='Shipping Cost v/s Profit|Category')
fig.show()
```

```python
# ols fit trendline and comparison
fig = px.scatter(data, x="Shipping Cost", y="Profit", color="Category", 
                 marginal_y="violin", marginal_x="box", trendline="ols",
                 opacity=0.15,
                 template="simple_white")
fig.show()
```

## Bar Chart

```python
data_bar = data.groupby('Category')['Profit'].sum()
data_bar = data_bar.to_frame()
data_bar
```

```python
data_bar.reset_index(level=0, inplace=True)
data_bar
```

```python
data_bar.iplot(kind='bar', x='Category', y='Profit',
             xTitle='Category', yTitle='Profit', 
              title='Category v/s Profit')
```

```python
data_bar = data.groupby(['Category','Sub-Category'])['Profit'].sum()
```

```python
data_bar=data_bar.to_frame()
```

```python
data_bar.reset_index(level=[0,1], inplace=True)
```

```python
data_bar
```

```python
fig = px.bar(data_bar, x='Sub-Category', y='Profit',
             hover_data=['Category','Sub-Category','Profit'], color='Category',
             labels={'Sub-Category':'Product Sub-Category'}, height=400)
```

```python
fig.show()
```
